### Data Cleaning (Replace this with more specific notes!)

```python
# After discussing the structure of the data and any problems that need to be
#   cleaned, perform those cleaning steps in the second part of this section.

#convert ScheduledDay and AppointmentDay to DateTime format 
# then create appointment booking date and the appointment date columns

df_noshow['AppointmentDate'] = pd.to_datetime(df_noshow['AppointmentDay']).dt.date
df_noshow['ScheduledDate'] = pd.to_datetime(df_noshow['ScheduledDay']).dt.date
df_noshow['AppointmentDataDay'] = pd.to_datetime(df_noshow['AppointmentDay']).dt.day_name()
```

```python
# Drop AppointmentDay and ScheduleDay
df_noshow.drop('AppointmentDay',axis=1 , inplace=True)
df_noshow.drop('ScheduledDay',axis=1 , inplace=True)
# new column for the waiting duration
df_noshow['DifferenceDays'] = df_noshow.AppointmentDate - df_noshow.ScheduledDate
```

```python
df_noshow.dtypes
```

```python
# Exclude the one record with the negative Age value
df_noshow = df_noshow[df_noshow.Age != -1]
```

```python
# Rename No-show column to be more user friendly 
df_noshow.rename(columns = {'No-show':'Show_Status'}, inplace = True)
```

```python
#Updating values in the previous column to be more convenient
df_noshow['Scholarship'] = df_noshow.Scholarship.astype('category')
df_noshow.Scholarship.cat.rename_categories(['No','Yes'], inplace = True)

df_noshow['Hipertension'] = df_noshow.Hipertension.astype('category')
df_noshow.Hipertension.cat.rename_categories(['No','Yes'], inplace = True)

df_noshow['Diabetes'] = df_noshow.Diabetes.astype('category')
df_noshow.Diabetes.cat.rename_categories(['No','Yes'], inplace = True)

df_noshow['Alcoholism'] = df_noshow.Alcoholism.astype('category')
df_noshow.Alcoholism.cat.rename_categories(['No','Yes'], inplace = True)

df_noshow['Handcap'] = df_noshow.Handcap.astype('category')
df_noshow.Handcap.cat.rename_categories(['No','handcap 1','handcap 2','handcap 3','handcap 4'], inplace = True)

df_noshow['SMS_received'] = df_noshow.SMS_received.astype('category')
df_noshow.SMS_received.cat.rename_categories(['No','Yes'], inplace = True)

df_noshow['Show_Status'] = df_noshow.Show_Status.astype('category')
df_noshow.Show_Status.cat.rename_categories(['No','Yes'], inplace = True)

df_noshow['AppointmentDataDay'] = df_noshow.AppointmentDataDay.astype('category')
df_noshow.AppointmentDataDay.cat.reorder_categories(['Friday', 'Tuesday', 'Monday', 'Wednesday', 'Thursday', 'Saturday'], 
                                    ordered = True, inplace = True)
```

```python
#convert Difference days to integer
# had an error at first with timedelta64[ns] 
#fixed with the help of this 
#https://stackoverflow.com/questions/33605514/how-to-extract-days-as-integers-from-a-timedelta64ns-object-in-python
df_noshow['DifferenceDays'] = (df_noshow.AppointmentDate - df_noshow.ScheduledDate).astype('timedelta64[D]').astype(int)
```

```python
# patients that have multiple appointments at the same day is duplictes
duplicated = df_noshow[['PatientId','AppointmentDate']].duplicated()
duplicated.sum()
```

```python
df_noshow = df_noshow[~duplicated]
```

## Exploratory Data Analysis

```python
# The percentage of No_show patients
no_show = df_noshow.groupby('Show_Status').count().loc['No'][0]
total_number = df_noshow.shape[0]
r= no_show/total_number
print(" No show cases = {}\n The total cases = {}\n The ratio= {}".format(no_show,total_number,r))
```

```python
#with the help of this https://matplotlib.org/stable/gallery/scales/log_bar.html#sphx-glr-gallery-scales-log-bar-py
# https://matplotlib.org/stable/gallery/lines_bars_and_markers/barchart.html#sphx-glr-gallery-lines-bars-and-markers-barchart-py
# a function to plot two bars side by side 
def plot_two_bars(df,column_name,x_labels=['',''],rotation=0):
    
    df_column = df.groupby(['Show_Status',column_name]).count() #grouping by column_name.
    bar_values=[df_column.PatientId.loc['No'],df_column.PatientId.loc['Yes']] #preparing the data  for two barplots
    figure ,ax = plt.subplots(2,1)
    width = .25 #width of bar
    bar_labels=['Show','No-Show']
    for i in range(2): # loop to plot two barplots side to side
        ax[0].bar(np.arange(0+i*width,2,1),bar_values[i],label=bar_labels[i],width=width)
        ax[0].set_ylim(0,max(np.append(bar_values[0].values,(bar_values[1].values)))*1.15) 
        ax[0].set_xlim(-0.25,1.5)
        ax[0].set_xticks([0.125,1.125])
        ax[0].set_xticklabels(x_labels,rotation = rotation,size=16)
        ax[0].set_ylabel('Number of cases',size=16)

    bar_ratio = bar_values[1]/(bar_values[0]+bar_values[1])
    ax[1].bar ([0.25,1.25],bar_ratio,label='Ratio',width=.25) #calculating Ratio
    ax[1].set_xlim(-.125,1.5)
    ax[1].set_xticks([0.25,1.25])
    ax[1].set_xticklabels(x_labels,size=16)
    ax[1].grid()
    ax[1].legend(loc='upper center',fontsize=14)
    ax[0].grid()
    ax[0].legend(fontsize=14)
    ax[0].set_title('Total cases based on {} '.format(column_name),size=15)
    ax[1].set_title('Ratio of no-show based on {}'.format(column_name),size=15)
    ax[1].set_ylabel('Ratio',size=15)
    ax[1].set_xlabel(column_name.upper(),size=15)
    ax[0].set_xlabel(column_name.upper(),size=15)

    figure.set_figheight(18)
    figure.set_figwidth(10)
    plt.show()
    return bar_ratio
 
```


### Research Question 1 (Is it related to one of the diseases?)

```python
plot_two_bars(df_noshow,'Diabetes',['Without \n Diabetes','With Diabetes'])
```

```python
plot_two_bars(df_noshow,'Alcoholism',['Without \n Alcoholism','With Alcoholism'])
```

```python
plot_two_bars(df_noshow,'Hipertension',['Without \n Hipertension','With Hipertension'])
```

##### There isn't  a significant relation between showing to an appointment and having a certain disease
